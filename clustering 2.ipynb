```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
df_norm1 = norm_func(data.iloc[:,1:]) # Normalizing the data
df_norm1
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
```

```python
z = linkage(df_norm1, method="single",metric="euclidean")

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(z,leaf_rotation=0.,leaf_font_size=8., )
plt.show()
```

```python
z = linkage(df_norm1, method="average",metric="euclidean") ## distance between the clusters , method= average 

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram( z,leaf_rotation=0.,  leaf_font_size=8.,)  
plt.show()
```

```python
z = linkage(df_norm1, method="complete",metric="euclidean") ## distance between the clusters , method= complete

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(z,leaf_rotation=0., leaf_font_size=8.,)  
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(df_norm1) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
data['clust']=cluster_labels 
data
```

```python
#kmeans clustering
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
stddata = pd.DataFrame(scaler.fit_transform(data.iloc[:,1:]))
stddata
```

```python
from sklearn.cluster import KMeans
import numpy as np
model1 = KMeans(n_clusters=5).fit(stddata)
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(stddata)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(stddata)
```

```python
y_kmeans

clf.labels_
```

```python
md=pd.Series(y_kmeans)  
data['clust']=md 
data
```

```python
data.iloc[:,1:].groupby(data.clust).mean()
```

```python
data.plot(x="Balance",y ="Bonus_miles",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans') 
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(stddata)
pred = kmeans.predict(stddata)
pred
```

```python
frame = pd.DataFrame(stddata)
frame['cluster'] = pred
frame['cluster'].value_counts()
```
